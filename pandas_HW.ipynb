# Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
ratings = pd.read_csv('small_ratings.csv')
ratings.head()
```

```python
data = ratings[['movieId', 'rating']]
len(data)
```

```python
# оставим только фильмы с рейтингом 5
max_rating = data[(data['rating']== 5)]
len(max_rating)
```

```python
# посчитаем количество по фильмам
max_rating['movieId'].value_counts().head(10)
```

```python
# другой способ

max_rating.groupby('movieId').count().sort_values('rating', ascending=False).head()
```

Посмотрим что это за фильмы

```python
movies = pd.read_csv('movies.csv')
```

```python
joined = max_rating.merge(movies, on='movieId', how='left')
joined.head()
```

```python
# топ фильмов по количеству лучших оценок
joined.groupby('title').count().sort_values('rating', ascending=False).head()
```

# Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.


```python
countries = pd.read_csv('power.csv')
countries.head()
```

```python
# для проверки как именно написано нахвание страны
countries[countries['country'].str.contains('la', case=False) ]['country'].unique()
```

```python
filtered_countries = countries[((countries['country']=='Latvia') | 
                               (countries['country']=='Lithuania') | 
                               (countries['country']=='Estonia')) & 
                               
                               ((countries['category']== 4)| 
                               (countries['category']== 12)| 
                               (countries['category']== 21)) &
                               
                               ((countries['year'] > 2005) & 
                               (countries['year'] < 2010)) &
                               
                               (countries['quantity'] > 0)]

filtered_countries.head()
```

```python
# для проверки
filtered_countries['country'].unique()
```

```python
filtered_countries.describe()
```

```python
print('Суммарное потребление стран Прибалтикит равно', filtered_countries['quantity'].sum()) 
```
